## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application2_df = application_df.drop(columns = ["EIN", "NAME"])
```

```python
application2_cat = application2_df.dtypes[application2_df.dtypes == "object"].index.tolist()
application2_df[application2_cat].nunique()
```

```python
# create dataframe of number of each type of application
count_types_df = pd.DataFrame(application2_df["APPLICATION_TYPE"].value_counts())
count_types_df
```

```python
#get the list of the top ten most common application types
most_common = application2_df["APPLICATION_TYPE"].value_counts()[:10]
most_common
```

```python
# create a list of application types (named application_types_to_replace) that lists those with fewer than 200 of that type
application_types_to_replace = []
for row in range (len(count_types_df)):
  if (count_types_df['APPLICATION_TYPE'][row] < 200):
    application_types_to_replace.append(count_types_df.index[row])

application_types_to_replace
```

```python
# Replace in dataframe
for app in range (len(application_types_to_replace)):
    application2_df['APPLICATION_TYPE'] = application2_df['APPLICATION_TYPE'].replace(application_types_to_replace[app],"Other")

# Check to make sure binning was successful
application2_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classify_count_df = pd.DataFrame(application2_df['CLASSIFICATION'].value_counts())
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
application2_df['CLASSIFICATION'].value_counts().loc[lambda x : x > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced; cutoff value of 50 selected
# use the variable name `classifications_to_replace`
classification_types_to_replace = []
for row in range (len(classify_count_df)):
  if (classify_count_df['CLASSIFICATION'][row] < 51):
    classification_types_to_replace.append(classify_count_df.index[row])


# Replace in dataframe
for cls in classification_types_to_replace:
    application2_df['CLASSIFICATION'] = application2_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application2_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_dummy_df = pd.get_dummies(application2_df, columns = ['INCOME_AMT', 'SPECIAL_CONSIDERATIONS', 'AFFILIATION', 'USE_CASE', 'ORGANIZATION', 'CLASSIFICATION', 'APPLICATION_TYPE'])
```

```python
application_dummy_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_dummy_df.IS_SUCCESSFUL.values
X = application_dummy_df.drop(columns = 'IS_SUCCESSFUL').values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 1, stratify = y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=77, activation="relu", input_dim=45))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="relu"))

# Check the structure of the model
nn_model.summary()


```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)

```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")

```

```python
# Export our model to HDF5 file
nn_model.save_weights('AlphabetSoupCharity_Optimization.h5')
```
